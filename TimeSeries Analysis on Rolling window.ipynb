# Visualising the Data

```python
%matplotlib inline
import seaborn
seaborn.set()
```

```python
a.plot()
plt.ylabel('Hourly Bicycle Count');
```

```python
weekely=a.resample('W').sum()
weekely.head()
```

```python
weekely.plot(style=['.-','--','-'])
plt.ylabel('Weekely bicycle count');
```

```python
a.plot(style=['.-','--','-'])
plt.ylabel('monthly bicycle count');
```

```python
daily=a.resample('D').sum()
daily.head()
```

```python
daily.plot(style=['.-','--','-'])
plt.ylabel('daily bicycle count');
```

```python
daily.rolling(120,center=True).std().plot(style=[':','--','-'])  # Mean + std
plt.ylabel('mean hourly count');
```

```python
daily.rolling(120,center=True).mean().plot(style=[':','--','-'])  # Mean + std
plt.ylabel('mean hourly count');
```

```python
daily.rolling(50,center=True,win_type='gaussian').sum(std=10).plot(style=[':','--','-']);
```

```python
by_time=a.groupby(a.index.time).sum()
by_time.head()
```

```python
plt.figure(figsize=(16,9))
hourly_ticks=4*60*60*np.arange(6)
by_time.plot(xticks=hourly_ticks,style=[':','--','-']);
```

```python
by_weekday=a.groupby(a.index.dayofweek).sum()
by_weekday.head()
```

```python
by_weekday.index=['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
by_weekday.plot(style=[':','--','-'])
```

```python
weekend=np.where(a.index.weekday<5,'Weekday','Weekend')
weekend[0:50]
```

```python
by_time=a.groupby([weekend,a.index.time]).mean()
by_time.head()
```

```python
by_time.tail()
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,8))
by_time.loc['Weekday'].plot(ax=ax[0],title='Weekdays',
                           xticks=hourly_ticks,style=[':','--','-'])
by_time.loc['Weekend'].plot(ax=ax[1],title='Weekends',
                           xticks=hourly_ticks,style=[':','--','-']);
```
